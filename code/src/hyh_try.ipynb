```python
import pandas as pd
from pathlib import Path

root = Path("/home/ye_canming/repos/novelties/ts/comp/AutoCute/output")
csv_files = sorted(root.glob("*.csv"))

# 读取并重命名列
dfs = []
for f in csv_files:
    stem = f.stem
    dfs.append(pd.read_csv(f).rename(columns=lambda c: f"{stem}_{c}"))

merged = pd.concat(dfs, axis=1)

# 每列降序排序
sorted_df = merged.apply(lambda col: col.sort_values(ascending=False).reset_index(drop=True))

# 拆成“奇数列”和“偶数列”两张表
odd_cols  = sorted_df.columns[::2]   # 第 0,2,4... 列
even_cols = sorted_df.columns[1::2]  # 第 1,3,5... 列

up  = sorted_df[odd_cols]
down = sorted_df[even_cols]


```
