```python
import pandas as pd

# 读入 CSV 文件
df_test = pd.read_csv("/home/ye_canming/repos/novelties/ts/comp/stork_predict_hyh/alpha158_test_lag.csv")
df_valid = pd.read_csv("/home/ye_canming/repos/novelties/ts/comp/stork_predict_hyh/alpha158_valid_lag.csv")
#df_train = pd.read_csv("/home/ye_canming/repos/novelties/ts/comp/stork_predict_hyh/alpha158_train_lag.csv")
```

清理test data

```python
# 确保 timestamp 列为 datetime 类型
df_test['timestamp'] = pd.to_datetime(df_test['timestamp'], errors='coerce')

# 筛选指定时间范围
start_date = pd.Timestamp('2025-04-26')
end_date = pd.Timestamp('2025-04-30')

df_test2 = df_test[(df_test['timestamp'] >= start_date) & (df_test['timestamp'] <= end_date)].reset_index(drop=True)

# 指定输出路径
output_file = "/home/ye_canming/repos/novelties/ts/comp/stork_predict_hyh/alpha158_test_lag2.csv"

# 保存为 CSV
df_test2.to_csv(output_file, index=False)

print(f"\n✅ df_test2 保存为 {output_file}\n")
print(df_test2.head())

```

清理validation data

```python
# 确保 timestamp 列为 datetime 类型
df_valid['timestamp'] = pd.to_datetime(df_valid['timestamp'], errors='coerce')

# 筛选 2025-04-20 ~ 2025-04-25 (含)
start_date = pd.Timestamp('2025-04-20')
end_date = pd.Timestamp('2025-04-25')

df_valid2 = df_valid[(df_valid['timestamp'] >= start_date) & (df_valid['timestamp'] <= end_date)].reset_index(drop=True)

# 指定输出路径
output_file = "/home/ye_canming/repos/novelties/ts/comp/stork_predict_hyh/alpha158_valid_lag2.csv"

# 保存为 CSV
df_valid2.to_csv(output_file, index=False)

print(f"\n✅ 已成功将筛选后的 df_valid2 保存为 {output_file}\n")
print(df_valid2.head())
```
